```python
l=0
for i in (df["Region"].unique()):
    print("Mean Happiness Score of",end=' ')
    print(i,"is",end=' ')
    k=z.get_group(i)
    print(k["Happiness Score"].mean())
    if (k["Happiness Score"].mean())>l:
        l=k["Happiness Score"].mean()
        m=i
```

```python
from sklearn import datasets, linear_model
lm = linear_model.LinearRegression()
x=df[["Economy (GDP per Capita)"]]
y=df[["Happiness Score"]]
lm.fit(x,y)
Z=np.linspace(0.5,1.0,100)
Z=Z.reshape(Z.size,1)
yhat=lm.predict(Z)
print(yhat)
```

```python
lm1=linear_model.LinearRegression()
Z1=df[["Freedom","Economy (GDP per Capita)","Health (Life Expectancy)"]]
lm1.fit(Z1,df["Happiness Score"])
```

```python
H_Value=lm1.predict(df[["Freedom","Economy (GDP per Capita)","Health (Life Expectancy)"]])
```

```python
Happiness_Score_list=df["Happiness Score"].tolist()
average1=0
for i in range(158):
    p=H_Value[i]-Happiness_Score_list[i]
    average1=average1+p
    print(p)
```
